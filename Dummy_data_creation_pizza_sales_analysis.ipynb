```python
import random
from datetime import datetime, timedelta
import pandas as pd

# Ingredients data
ingredients_data = {
    "Flour": 2,
    "Tomato sauce": 1,
    "Mozzarella cheese": 3,
    "Pepperoni": 4,
    "Chicken": 5,
    "Vegetables": 2,
    "Paneer": 3,
    "Spices": 1,
    "Seafood": 6
}

# Generating data for each month from 2021 to 2023
start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 3, 31)
month_count = (end_date.year - start_date.year) * 12 + end_date.month - start_date.month + 1

data = []
for single_date in (start_date + timedelta(n*365/12) for n in range(month_count)):
    for ingredient, cost in ingredients_data.items():
        month = single_date.strftime("%B")
        year = single_date.year
        data.append([month, year, ingredient*30, cost])

# Convert the data to a DataFrame
df = pd.DataFrame(data, columns=['Month', 'Year', 'Ingredient', 'Cost'])

# Save the DataFrame to a CSV file
df.to_csv('ingredients_cost_monthly.csv', index=False)
print("Data saved successfully to 'ingredients_cost_monthly.csv'.")
```
